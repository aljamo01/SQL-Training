### Show the country name and life expectancy of the 5 countries with the longest life expectancy.

```python
%%sql
SELECT name,lifeexpectancy
FROM country
order by lifeexpectancy DESC
limit 5 
;
```


### Show the country name, population, region, continent and gnp of the 8 countries with the largest gnp.

```python
%%sql
SELECT name, population, region, continent, gnp
FROM country
order by gnp DESC
limit 8
;
```
